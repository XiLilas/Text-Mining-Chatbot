***Cyk and grammar***

```python
from cyk import CYK

def to_cyk_input(tagged_sentence):
    words, pos_choices = [], []
    for w, tags in tagged_sentence:
        if "PUNCT" not in tags:
            words.append(w)
            pos_choices.append(tags)
    return words, pos_choices

G = [
    # S is axiom
    ("S", ("NP", "VP")), 
    ("S", ("S", "S")), 
    
    
    # non terminal rules
    ("NP", ("DET", "NOUN")),
    ("NP", ("DET", "PROPN")),
    ("NP", ("DET", "NP")),
    ("NP", ("NOUN",)),
    ("NP", ("PROPN",)),
    ("NP", ("NUM",)),
    ("NP", ("DET",)), #this in 'astronomy'
    ("NP", ("NP", "PP")),
    ("NP", ("NP", "NP")),
    ("NP", ("NP", "NUM")),
    ("NP", ("ADJ", "NOUN")), #under high ambient temperature
    ("NP", ("ADJ", "NP")),

    ("PP", ("ADP", "NP")),

    ("VP", ("VERB",)),
    ("VP", ("AUX", "VERB")),
    ("VP", ("AUX", "NP")), #this is a type-Ia
    ("VP", ("VERB", "NP")),
    ("VP", ("VERB", "PP")),
    ("VP", ("VERB", "ADJ")),
    ("VP", ("VP", "PP")),
    ("VP", ("VP", "VP")),
    ("VP", ("VP", "SCONJ")),

    # terminal (if using pos directly)
    ("DET", ("DET",)),
    ("VERB", ("VERB",)),
    ("NOUN", ("NOUN",)),
    ("PROPN", ("PROPN",)),
    ("AUX", ("AUX",)),
    ("ADP", ("ADP",)),
    ("ADJ", ("ADJ",)),
    ("NUM", ("NUM",)),
    ("SCONJ", ("SCONJ",))
]

'''
tag = POS_sentences["s1"]
words, pos_choices = to_cyk_input(tag)

print("words : ",  words)
print("pos choices : ", pos_choices)

cyk = CYK(G)
trees = cyk(pos_choices, words)

print(trees)

for i, t in enumerate(trees[:10], 1):
        print(f"\nTree {i}:")
        print(t)
'''
```

```python
def parse_and_show(sentence_id, grammar, max_trees=5):
    """
    Run CYK parsing for a given sentence and grammar,
    then print words, POS choices and resulting trees.
    """

    print("\n" + "=" * 60)
    print(f"Sentence ID: {sentence_id}")

    # 1. Get tagged sentence
    tagged_sentence = POS_sentences[sentence_id]

    # 2. Convert to CYK input (remove punctuation)
    words, pos_choices = to_cyk_input(tagged_sentence)

    print("Words:")
    print(words)

    print("POS choices:")
    print(pos_choices)

    # 3. Run CYK
    cyk = CYK(grammar)
    trees = cyk(pos_choices, words)

    # 4. Print results
    print(f"\nNumber of trees: {len(trees)}")

    if not trees:
        print("No valid parse trees found.")
        return

    for i, t in enumerate(trees[:max_trees], 1):
        print(f"\nTree {i}:")
        print(t)
```

```python
sentence_ids = ["s1", "s2", "s3", "s4", "biology", "astronomy"]

for i, sid in enumerate(sentence_ids, 1):
    print(f"\n########## Sentence {i}: {sid} ##########")
    parse_and_show(sid, G)
```
